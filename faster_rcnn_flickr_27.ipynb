```
import pandas as pd

data = pd.read_csv('flickr_logos_27_dataset/flickr_logos_27_dataset_training_set_annotation.txt', sep=" ", header=None)
```

```
data=data.drop([2,7],axis=1)
data
```

```
data.columns = [0,3,4,5,6,7]
data
```

```
import cv2
heightlist=[]
widthlist=[]
for lala in data.iterrows():
    print(lala[1][0])
    image= cv2.imread('flickr_logos_27_dataset/flickr_logos_27_dataset_images/'+str(lala[1][0]))
    print(image.shape)
    H,W,_=image.shape
    heightlist.append(H)
    widthlist.append(W)
```

```
data.insert(2, "width", widthlist, True)
data.insert(1, "height", heightlist, True)
data 
```

```
data.columns= ['filename','height','class','width','xmin','ymin','xmax','ymax']
```

```
data=data[['filename','height','width','class','xmin','ymin','xmax','ymax']]
```

```
data.to_csv('flickr_logos_27_dataset/annotations.csv')
```

```
import os
os.mkdir('flickr_logos_27_dataset/train')
os.mkdir('flickr_logos_27_dataset/test')

os.listdir()
```

```
images_path= os.listdir('flickr_logos_27_dataset/flickr_logos_27_dataset_images')
for i,path in enumerate(images_path):
  images_path[i]='flickr_logos_27_dataset/flickr_logos_27_dataset_images/'+str(images_path[i])
images_path
```

```
splitorder=int(len(images_path)*0.7)
splitorder
```

```
import  random
random.shuffle(images_path)
```

```
from shutil import copyfile
for path in images_path[:splitorder]:
  naam= path.split('/')[-1]
  copyfile(path, 'flickr_logos_27_dataset/train/'+naam)
for path in images_path[splitorder:]:
  naam= path.split('/')[-1]
  copyfile(path, 'flickr_logos_27_dataset/test/'+naam)
```

```
image_path_train = os.listdir('flickr_logos_27_dataset/train')
```

```
df = pd.read_csv("flickr_logos_27_dataset/annotations.csv")
```

```
df_train= []
df_test= []
for index,row in df.iterrows():
  if row['filename'] in image_path_train:
    df_train.append(row)
  else:
    df_test.append(row)
```

```
df_train= pd.DataFrame(df_train)
df_test= pd.DataFrame(df_test)

df_train= pd.DataFrame(df_train).drop(['Unnamed: 0'],axis=1)
df_test= pd.DataFrame(df_test).drop(['Unnamed: 0'],axis=1)
df_test
```

```
df_test.to_csv('flickr_logos_27_dataset/test_annotations.csv')
df_train.to_csv('flickr_logos_27_dataset/train_annotations.csv')
```
